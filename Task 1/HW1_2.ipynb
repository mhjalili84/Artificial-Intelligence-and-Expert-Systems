```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from google.colab import drive
drive.mount('/content/drive')

```

```python
df = pd.read_csv("/content/drive/MyDrive/dataset/AIES_HW_1/Dataset_II.csv", header=None)
data = np.array(df)
X = data[1].reshape(-1,1)  # Feature (independent variable)
y = data[0].reshape(-1,1)     # Target variable (dependent variable)
```

```python
# Extracting the features and target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Define the degree of the polynomial
degree = 2

# Transforming features into polynomial features
poly_features = PolynomialFeatures(degree=degree)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.fit_transform(X_test)

# Fitting the polynomial regression model
model = LinearRegression()
model.fit(X_poly_train, y_train)

# Predicting train
y_pred = model.predict(X_poly_train)

# Predicting on the test data
# Calculating Mean Squared Error (MSE)
mse = mean_squared_error(y_train, y_pred)
# Calculating Mean Absolute Error (MAE)
mae = mean_absolute_error(y_train, y_pred)

print('MSE = ', mse)
print('MAE = ', mae)
# Predicting test
y_pred = model.predict(X_poly_test)


# Predicting on the test data
# Calculating Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
# Calculating Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(model.coef_,model.intercept_)
print('MSE = ', mse)
print('MAE = ', mae)
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Regression for Test Data')
plt.title('Polynomial Regression (order = 3)')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extracting the features and target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Define the degree of the polynomial
degree = 1

# Transforming features into polynomial features
poly_features = PolynomialFeatures(degree=degree)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.fit_transform(X_test)


# Transforming features into Sin and Cos features
X_sin_train = np.sin(X_train)
X_cos_train = np.cos(X_train)
X_sin_test = np.sin(X_test)
X_cos_test = np.cos(X_test)

feature_train = np.concatenate((X_poly_train,X_sin_train,X_cos_train),axis = 1)
feature_test = np.concatenate((X_poly_test,X_sin_test,X_cos_test),axis = 1)
# Fitting the polynomial regression model
model = LinearRegression()
model.fit(feature_train, y_train)
# Predicting train
y_pred = model.predict(feature_train)

# Predicting on the test data
# Calculating Mean Squared Error (MSE)
mse = mean_squared_error(y_train, y_pred)
# Calculating Mean Absolute Error (MAE)
mae = mean_absolute_error(y_train, y_pred)

print('MSE = ', mse)
print('MAE = ', mae)

# Predicting test
y_pred = model.predict(feature_test)

# Predicting on the test data
# Calculating Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
# Calculating Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(model.coef_,model.intercept_)
print('MSE = ', mse)
print('MAE = ', mae)
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Regression for Test Data')
plt.title('Polynomial and Sin Regression')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
